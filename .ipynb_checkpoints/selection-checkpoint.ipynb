```python
anchorOffset = 5
focusOffset = 2

def get_element_text(element):
    if element['tagName'] == '':
        text = element['textContent']
    elif element['tagName'] == 'b':
        text = '<b>' + element['textContent'] + '</b>'
    return text

def get_anchor_is_focus_text(element, anchorOffset, focusOffset):
    if element['tagName'] == '':
        text_bl = element['textContent'][:anchorOffset]
        text_sl = element['textContent'][anchorOffset:focusOffset]
        text_el = element['textContent'][focusOffset:]
    elif element['tagName'] == 'b':
        text_bl = '<b>' + element['textContent'][:anchorOffset] + '</b>'
        text_sl = '<b>' + element['textContent'][anchorOffset:focusOffset] + '</b>'
        text_el = '<b>' + element['textContent'][focusOffset:] + '</b>'
    return text_bl, text_sl, text_el
    
    
def get_anchor_text(element, anchorOffset):
    anchor_text_bl = element['textContent'][:anchorOffset]
    anchor_text_sl = element['textContent'][anchorOffset:]
    
    if element['tagName'] == '':
        return anchor_text_bl, anchor_text_sl
    elif element['tagName'] == 'b':
        return '<b>' + anchor_text_bl + '</b>', '<b>' + anchor_text_sl + '</b>'
    
   
def get_focus_text(element, focusOffset):
    focus_text_sl = element['textContent'][:focusOffset]
    focus_text_el = element['textContent'][focusOffset:]
    
    if element['tagName'] == '':
        return focus_text_sl, focus_text_el
    elif element['tagName'] == 'b':
        return '<b>' + focus_text_sl + '</b>', '<b>' + focus_text_el + '</b>'    
    
def get_html_text(element, anchorOffset, focusOffset):

    def get_full_text_lists(element, begining_of_selection_found = False, end_of_selection_found = False):
        begining_list = []
        selection_list = []
        end_list = []

        if element['isInSelection']:
            begining_of_selection_found = True
        if begining_of_selection_found and not element['isInSelection']:
            end_of_selection_found = True
        if element['tagName'] == '' or element['tagName'] == 'b':
            if not begining_of_selection_found:
                begining_list.append(element)
            elif not end_of_selection_found:
                selection_list.append(element)
            else:
                end_list.append(element)
        else:
            for el in element['children']:
                bl, sl, el, begining_of_selection_found, end_of_selection_found = get_full_text_lists(el, begining_of_selection_found, end_of_selection_found)
                begining_list.extend(bl)
                selection_list.extend(sl)
                end_list.extend(el)
            
        return begining_list, selection_list, end_list, begining_of_selection_found, end_of_selection_found

    bl, sl, el, bfnd, efnd = get_full_text_lists(element)

    text = ''
    for element in bl:
        text += get_element_text(element)
    print(f"The length of sl is : {len(sl)}")
    if len(sl) == 0:
        pass
    elif len(sl) == 1:
        text_bl, text_sl, text_el = get_anchor_is_focus_text(sl[0], anchorOffset, focusOffset)
        text += text_bl
        text += '<mark>' + text_sl + '</mark>'
        text += text_el
    elif len(sl) > 1:
        anchor_text_bl, anchor_text_sl = get_anchor_text(sl[0], anchorOffset)
        print("coucou", text)
        text += anchor_text_bl + '<mark>' + anchor_text_sl
        print("cucu", text)
        for element in sl[1:-1]:
            text += get_element_text(element)
        focus_text_sl, focus_text_el = get_focus_text(sl[-1], focusOffset)
        text += focus_text_sl + '</mark>' + focus_text_el
    for element in el:
        text += get_element_text(element)
    
    return text
```

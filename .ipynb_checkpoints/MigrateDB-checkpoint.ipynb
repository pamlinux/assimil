```python
stmt_old = select(databaseold.LessonSession)
with Session(engine_old) as session_old:
    for lesson_session in session_old.scalars(stmt_old):
        print(lesson_session)
```

```python
with Session(engine_old) as session_old:
    for lesson_session in session_old.scalars(stmt_old):
        print(lesson_session)
        errored_sentences = []
        for sentence in lesson_session.errored_sentences:
            errored_sentences.append(database.MarkedSentence(
                sentence = sentence.sentence,
                comment = sentence.comment,
                line = sentence.line
            ))
        pam.sessions.append(database.LessonSession(
            user=pam,
            lesson = lesson_session.lesson,
            errors_number = lesson_session.errors_number,
            date_time = lesson_session.date_time,
            errored_sentences = errored_sentences
        ))
    session_new.commit()
```

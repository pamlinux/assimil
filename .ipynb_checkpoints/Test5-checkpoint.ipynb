```python
from sqlalchemy.orm import Session
from sqlalchemy import select, and_
from maintenance import database as db_mnt

def has_dash_dialogue(sentence):
    for k, c in enumerate(sentence):
        if c.isdigit():
            pass
        elif c.isalpha():
            return False, sentence[k:]
        elif c == '-':
            return True, sentence[k+2:]
    raise
            
def get_french_paragraphs(lesson_nb, level = 1):
    
    fn = f"lessons/new/L{str(lesson_nb).zfill(3)}.txt"
    lines = open(fn).readlines()
    paragraphs = [has_dash_dialogue(l.replace("\n", '')) for l in lines]
    return paragraphs

def get_spanish_paragraphs(lesson_nb):
    para_dict = {}
    db_mnt_stmt = select(db_mnt.Paragraph).where(db_mnt.Paragraph.lesson_nb == lesson_nb)
    with Session(db_mnt.engine) as db_mnt_session:
        for entry in db_mnt_session.scalars(db_mnt_stmt):
            para_dict[entry.line_nb] = (entry.paragraph, entry.section, entry.line_nb)
    lines_number = len(para_dict.keys())
    para_list = []
    for k in range(lines_number):
        para_list.append(para_dict[k][0])

    paragraphs = []
    
    for k, p in enumerate(para_list):
        if p[:3] == "S00":
            para = p[10:]
        elif p[:3] == "T00":
            para = p[14:]
        elif p[:4] == "ST00":
            para = p[15:]
        elif p[:2] == "ST":
            para = p[5:]
        elif p[0] == "N":
            pos = p.find('-')
            para = p[pos+1:]
        else:
            para = p[4:]
        paragraphs.append((para, para_dict[k][1], para_dict[k][2]))
    return paragraphs
 
```

```python
def get_french_paragraph_elements(paragraph):
    dash_found = False
    p = ""
    number = ""
    for i, c in enumerate(paragraph):
        if c.isdigit():
            number += c
        elif c == '-':
            dash_found = True
            index = i + 1
            break
        elif c == ' ':
            index = i + 1
        else:
            index = i
            break
            
    if number:
        number = int(number)
    return number, dash_found, paragraph[index:]
    
            
```

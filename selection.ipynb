```python
def mark_anchor_and_focus_nodes(element):
    global begining_of_selection_found 
    begining_of_selection_found = False
    end_of_selection_found = False
    global previous_element
    previous_element = None
     
    def mark_node(element):
        global previous_element
        global begining_of_selection_found
        if element['isInSelection'] and not begining_of_selection_found: #implies it's the anchor element
            begining_of_selection_found = True
            element['status'] = 'isAnchor'
            previous_element = element
        elif element['isInSelection'] and begining_of_selection_found:
            element['status'] = 'isInSelection'
            previous_element = element
        elif not element['isInSelection'] and previous_element and previous_element['status'] == 'isAnchor':
            element['status'] = ''
            previous_element['status'] = 'isAnchorAndFocus'
        elif not element['isInSelection'] and previous_element and previous_element['status'] == 'isInSelection':
            previous_element['status'] = 'isFocus'
            element['status'] = ''
            previous_element = element
        else:
            element['status'] = ''
            previous_element = element
        for child in element['children']:
            mark_node(child)
           
            
    mark_node(element)
            
```

```python
from html.parser import HTMLParser

class MarkedLinesParser(HTMLParser):
    def analyze_lesson(self, lesson):
        self.sentences = []
        self.marked_lines = []
        self.marked_lines_numbers = {}
        self.current_line_number = 0
        self.feed(lesson)

    def handle_starttag(self, tag, attrs):
        if tag == "p":
            self.current_sentence = ""
        elif tag == "mark":
            if not self.current_line_number in self.marked_lines_numbers:
                self.marked_lines_numbers[self.current_line_number] = True
            self.current_sentence += f"<{tag}>"
        else:
            self.current_sentence += f"<{tag}>"
        

    def handle_endtag(self, tag):
        if tag == "p":
            self.sentences.append(self.current_sentence.strip())
            self.current_line_number += 1
        else:
            self.current_sentence += f"</{tag}>"

    def handle_data(self, data):
        if hasattr(self, "current_sentence"):
            self.current_sentence += data
    
    def get_marked_lines_numbers(self):
        return self.marked_lines_numbers

    def get_sentences(self):
        return self.sentences
    
    
    
```

```python
import datetime
import re

def parse_srt_time_range(timestamp: str):
    """
    Converts an SRT timestamp into start_time and end_time.

    Example:
    Input: "00:01:23,456 --> 00:01:25,789"
    Output: (datetime.time(0, 1, 23, 456000), datetime.time(0, 1, 25, 789000))
    """
    pattern = r"(\d{2}):(\d{2}):(\d{2}),(\d{3})\s*-->\s*(\d{2}):(\d{2}):(\d{2}),(\d{3})"
    match = re.match(pattern, timestamp)

    if not match:
        raise ValueError(f"Invalid timestamp format: {timestamp}")

    h1, m1, s1, ms1, h2, m2, s2, ms2 = map(int, match.groups())

    start_time = datetime.time(h1, m1, s1, ms1 * 1000)
    end_time = datetime.time(h2, m2, s2, ms2 * 1000)

    return start_time, end_time
```

```python
import re

def load_vtt_subtitles(vtt_file):
    with open(vtt_file, "r", encoding="utf-8") as f:
        content = f.read()
    
    print("in load_subtitles file : ", vtt_file)

    # Supprimer le BOM si présent
    if content.startswith('\ufeff'):
        content = content[1:]

    # Supprimer la ligne WEBVTT si elle est présente
    lines = content.strip().split("\n")
    if lines[0].strip().upper() == "WEBVTT":
        lines = lines[1:]  # Ignorer la première ligne
    
    # Regrouper les lignes par blocs séparés par une ligne vide
    blocks = re.split(r"\n\s*\n", "\n".join(lines))
    subtitles = []

    for block in blocks:
        lines = block.strip().split("\n")
        if len(lines) < 2:
            continue  # Ignorer les blocs trop courts
        
        # Vérifier si la première ligne est un timestamp
        if "-->" in lines[0]:
            timestamp = lines[0].strip()
            text = "\n".join(lines[1:]).strip()
        elif len(lines) > 1 and "-->" in lines[1]:  # Si la première ligne est un ID, alors le timestamp est à la deuxième ligne
            timestamp = lines[1].strip()
            text = "\n".join(lines[2:]).strip()
        else:
            continue  # Ignorer les blocs mal formatés

        subtitles.append({
            "timestamp": timestamp,
            "text": text,
        })

    return subtitles
```

```python
import re

def load_srt_subtitles(srt_file):
    with open(srt_file, "r", encoding="utf-8") as f:
        content = f.read()
    
    print("in load_srt_subtitles file : ", srt_file)

    # Supprimer le BOM si présent
    if content.startswith('\ufeff'):
        content = content[1:]

    # Séparer les blocs de sous-titres
    blocks = re.split(r"\n\s*\n", content.strip())
    subtitles = []

    for block in blocks:
        lines = block.strip().split("\n")
        if len(lines) < 3:
            continue  # Ignorer les blocs trop courts
        
        # La deuxième ligne est normalement le timestamp
        timestamp_line = lines[1].strip()
        
        # Vérifier le format du timestamp (doit contenir "-->")
        if "-->" not in timestamp_line:
            continue

        # Récupérer le texte
        text = "\n".join(lines[2:]).strip()

        subtitles.append({
            "timestamp": timestamp_line,
            "text": text,
        })

    return subtitles
```

```python
import re

def fix_srt_timestamp_format(input_filename, output_filename):
    """
    Fixes the timestamp format in an SRT file by replacing commas with periods.

    Args:
        input_filename (str): The name of the input SRT file with the incorrect format.
        output_filename (str): The name of the output SRT file with the corrected format.
    """
    try:
        with open(input_filename, 'r', encoding='utf-8') as input_file, \
             open(output_filename, 'w', encoding='utf-8') as output_file:
            for line in input_file:
                # Search for timestamp lines in the format "HH:MM:SS,ms --> HH:MM:SS,ms"
                timestamp_match = re.match(r'(\d{2}:\d{2}:\d{2}),(\d{3}) --> (\d{2}:\d{2}:\d{2}),(\d{3})', line)
                if timestamp_match:
                    # Rebuild the line with periods instead of commas
                    hours1, milliseconds1, hours2, milliseconds2 = timestamp_match.groups()
                    corrected_line = f"{hours1}.{milliseconds1} --> {hours2}.{milliseconds2}\n"
                    output_file.write(corrected_line)
                else:
                    # Write other lines as they are
                    output_file.write(line)
        print(f"The file '{input_filename}' has been corrected and saved as '{output_filename}'.")

    #except FileNotFoundError:
    #    print(f"Error: The file '{input_filename}' was not found.")
    except Exception as e:
        print(f"An error occurred while processing the file: {e}")
        
```

```python
import re

def timestamp_to_seconds(timestamp_str):
    """
    Converts an SRT timestamp string (H+:MM:SS.mmm) to total seconds,
    handling cases with or without hours.

    Args:
        timestamp_str (str): The SRT timestamp string in the format
                             [H+:]MM:SS.mmm.

    Returns:
        float: The total time in seconds, or None if the format is invalid.
    """
    match_with_hours = re.match(r'(\d+):(\d{2}):(\d{2})\.(\d{3})', timestamp_str)
    match_without_hours = re.match(r'(\d{2}):(\d{2})\.(\d{3})', timestamp_str)

    if match_with_hours:
        hours, minutes, seconds, milliseconds = map(int, match_with_hours.groups())
        total_seconds = (hours * 3600) + (minutes * 60) + seconds + (milliseconds / 1000)
        return total_seconds
    elif match_without_hours:
        minutes, seconds, milliseconds = map(int, match_without_hours.groups())
        total_seconds = (minutes * 60) + seconds + (milliseconds / 1000)
        return total_seconds
    else:
        return None

if __name__ == '__main__':
    timestamp1 = "00:00:00.000"
    seconds1 = timestamp_to_seconds(timestamp1)
    print(f"'{timestamp1}' is equal to {seconds1} seconds.")

    timestamp2 = "01:30:05.250"
    seconds2 = timestamp_to_seconds(timestamp2)
    print(f"'{timestamp2}' is equal to {seconds2} seconds.")

    timestamp3 = "38:45.345"
    seconds3 = timestamp_to_seconds(timestamp3)
    print(f"'{timestamp3}' is equal to {seconds3} seconds.")

    timestamp4 = "00:05.123"
    seconds4 = timestamp_to_seconds(timestamp4)
    print(f"'{timestamp4}' is equal to {seconds4} seconds.")

    invalid_timestamp = "00:10:50"
    seconds5 = timestamp_to_seconds(invalid_timestamp)
    print(f"'{invalid_timestamp}' is equal to {seconds5}")

    timestamp_with_comma = "00:00:05,123"
    seconds6 = timestamp_to_seconds(timestamp_with_comma)
    print(f"'{timestamp_with_comma}' is equal to {seconds6}")
```

```python
import datetime

def time_to_seconds(time_object: datetime.time) -> int:
    """
    Converts a datetime.time object to the total number of seconds since midnight.

    Args:
        time_object: The datetime.time object.

    Returns:
        The total number of seconds as an integer.
    """
    total_seconds = (time_object.hour * 3600) + (time_object.minute * 60) + time_object.second
    return total_seconds

if __name__ == '__main__':
    time1 = datetime.time(0, 0, 0)
    seconds1 = time_to_seconds(time1)
    print(f"{time1} is {seconds1} seconds since midnight.")

    time2 = datetime.time(1, 30, 5)
    seconds2 = time_to_seconds(time2)
    print(f"{time2} is {seconds2} seconds since midnight.")

    time3 = datetime.time(12, 45, 30)
    seconds3 = time_to_seconds(time3)
    print(f"{time3} is {seconds3} seconds since midnight.")
```

```python
from mlconjug3 import Conjugator

conjugator = Conjugator(language='es')

def find_verb_form(form):
    possible_matches = []

    for verb in conjugator.verbs:  # tous les verbes connus du modèle
        conjugation = conjugator.conjugate(verb)
        for mood, tenses in conjugation.iteritems():
            for tense, persons in tenses.items():
                for person, value in persons.items():
                    if value == form:
                        possible_matches.append({
                            "infinitive": verb,
                            "mood": mood,
                            "tense": tense,
                            "person": person
                        })

    return possible_matches

results = find_verb_form("venga")
for r in results:
    print(r)
```
